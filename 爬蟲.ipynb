```python
def rename(link2): #將資料夾MP4全改名,運行前將無關MP4移走
    import glob
    import os
    os.chdir(link2)

    mp4 = glob.glob("*.mp4") #根據副檔名排序
    mp4.sort()

    n = 1          # 設定名稱從 1 開始
    for i in mp4:
        os.rename(i, f'{n:03d}.mp4')   # 改名時，使用字串格式化的方式進行三位數補零
        n = n + 1    # 每次重複時將 n 增加 1

```
